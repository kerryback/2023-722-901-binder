```python
import pandas as pd
```

```python
df = pd.read_csv('./crude_futures.csv')
df = df.dropna()
```

```python
import matplotlib.pyplot as plt

t = df['date']
data1 = df['settle']
data2 = df['settle'] * 3

fig, ax = plt.subplots(figsize=(15, 5))

# Create a second y-axis
ax.plot(t, data1, label='Crude Future')
ax.set_ylabel('Crude Future', color='red')
ax.plot(t, data2, label='Levered Crude Future')
ax.set_ylabel('Price $')
ax.tick_params('y')

# Add legend and titles
plt.xlabel('Time')
plt.title('Crude Oil Futures')
plt.legend()
plt.show()
```
