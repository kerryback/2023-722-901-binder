```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv('./crude_futures.csv')
df = df.dropna()
```

```python
def futures_code_to_date(code):
    month_code = code[-5]
    year_code = code[-4:]

    month_dict = {
        'F': '01', 'G': '02', 'H': '03', 'J': '04', 'K': '05', 'M': '06',
        'N': '07', 'Q': '08', 'U': '09', 'V': '10', 'X': '11', 'Z': '12'
    }
    month = month_dict[month_code]
    return f'{year_code}-{month}-01'
```

```python
futures_code_to_date('CME_CLF1984')
```

```python
df['exparation_date'] = [futures_code_to_date(c) for c in df['code'].values.tolist()]
```

```python
df = df[df['date'] == '2012-01-31']
```
